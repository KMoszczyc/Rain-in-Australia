Temperatures from april, may and june seem like they come from normal distribution..

```python
from scipy import stats
# groups are not dependent
# 3. basic statistics

# normality test
stat, p = stats.shapiro(sydney_april['Temp3pm'])
print('april',stat, p)

stat, p = stats.shapiro(sydney_may['Temp3pm'])
print('may',stat, p)

stat, p = stats.shapiro(sydney_june['Temp3pm'])
print('june',stat, p)
# quite normal
print()

# test wariancji - leven'a
stat, p = stats.levene(sydney_april['Temp3pm'], sydney_may['Temp3pm'], sydney_june['Temp3pm'])
print('test wariancji: ',stat, p)
# wariancja nie ok
#  First, there's a rule of thumb that the ANOVA is robust to heterogeneity of variance so long as the largest variance is not more than 4 times the smallest variance. 
var1 = np.var(sydney_april['Temp3pm'])
var2 = np.var(sydney_may['Temp3pm'])
var3 = np.var(sydney_june['Temp3pm'])
print('variances:', var1, var2, var3)
# Variances seem to be close to each other, we can do ANOVA



# h0: Temp3pm is the same in april, may, and june
# h1: Temp3pm are not the same in april, may, and june

# AVONA test, oneway
stat, p = stats.f_oneway(sydney_april['Temp3pm'], sydney_may['Temp3pm'], sydney_june['Temp3pm'])
print('ANOVA test',stat, p)

# h0 is rejected, h1 is accepted, means of Temp3pm in april, may, june are not the same
```

```python
import numpy as np 
# 4. checking means

print('means: ', 'april:', np.mean(sydney_april['Temp3pm']), 'may:',np.mean(sydney_may['Temp3pm']), 'june:', np.mean(sydney_june['Temp3pm']))

# a, b, alternative='greater' means a < b
# different variances so twelch test
# april - may
stat, p = stats.ttest_ind(sydney_april['Temp3pm'], sydney_may['Temp3pm'], alternative='greater', equal_var=False)
print('april - may',stat, p)

# may - june
stat, p = stats.ttest_ind(sydney_may['Temp3pm'], sydney_june['Temp3pm'], alternative='greater', equal_var=False)
print('may - june',stat, p)

#  april - june
stat, p = stats.ttest_ind(sydney_april['Temp3pm'], sydney_june['Temp3pm'], alternative='greater', equal_var=False)
print('april - june',stat, p)

# the biggest temperature in Sydney at 3pm occurs in April!
```

```python
import statsmodels.api as sm
import statsmodels.stats.descriptivestats as smsd
# 5
mean_april = np.mean(sydney_april['Temp3pm'])
mean_may = np.mean(sydney_may['Temp3pm'])
mean_june = np.mean(sydney_june['Temp3pm'])

print('means: ', mean_april, mean_may, mean_june)


#  april
stat, p = stats.ttest_1samp(sydney_april['Temp3pm'], mean_april*1.1, alternative='greater')
print('april',stat, p)

#  may
stat, p = stats.ttest_1samp(sydney_may['Temp3pm'], mean_may*1.1, alternative='greater')
print('may',stat, p)

#  june
stat, p = stats.ttest_1samp(sydney_june['Temp3pm'], mean_june*1.1, alternative='greater')
print('june',stat, p)
```

```python
# 6 przedziały ufności
alpha1=0.05
alpha2=0.01

n1 = len(sydney_april['Temp3pm'])
n2 = len(sydney_may['Temp3pm'])
n3 = len(sydney_june['Temp3pm'])
print(n1, n2, n3)

std1 = np.std(sydney_april['Temp3pm'])
std2 = np.std(sydney_may['Temp3pm'])
std3 = np.std(sydney_june['Temp3pm'])
print('stds:', std1, std2, std3)



# as n>30 we will use normal distribution
# N(0,1) 1-alpha/2 
z1 = stats.norm.ppf(0.975, 0, 1)
z2 = stats.norm.ppf(0.995, 0, 1)

print('norm quantile 0.975', z1)
print('norm quantile 0.995', z2)
print()

print('means confidence intervals 95%')
print('april:', '[', mean_april - z1 * std1/np.sqrt(n1), mean_april + z1 * std1/np.sqrt(n1),']')
print('may:', '[', mean_may - z1 * std2/np.sqrt(n2), mean_may + z1 * std2/np.sqrt(n2),']')
print('june:', '[', mean_june - z1 * std3/np.sqrt(n3), mean_june + z1*std3/np.sqrt(n3),']')
print()

print('means confidence intervals 99%')
print('april:', '[', mean_april - z2 * std1/np.sqrt(n1), mean_april + z2 * std1/np.sqrt(n1),']')

print('may:', '[', mean_may - z2 * std2/np.sqrt(n2), mean_may + z2 * std2/np.sqrt(n2),']')

print('june', '[', mean_june - z2 * std3/np.sqrt(n3), mean_june + z2*std3/np.sqrt(n3),']')
```
